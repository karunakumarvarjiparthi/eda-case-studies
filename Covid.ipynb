```python
Highest = df.groupby('Name of State / UT') ['Total Confirmed cases'].max().sort_values(ascending=False).head()
Highest
```

```python
top5 = df.groupby('Name of State / UT') ['Total Confirmed cases'].max().sort_values(ascending=False).head()

plt.bar(top5.index, top5.values, label="Confirmed cases")


plt.title("COVID Cases by State")
plt.ylabel("Number of Cases")
plt.xlabel("State")
plt.legend(title="Case Type")
plt.show()
```

```python

top5 = df.groupby('Name of State / UT')['Death'].max().sort_values(ascending=False).head()
plt.bar(top5.index, top5.values, label="Death cases")

plt.title("COVID Death Cases")
plt.ylabel("Number of Cases")
plt.xlabel("State")
plt.legend(title="Case Type")
plt.show()
```

```python
top5 = df.groupby('Name of State / UT')['Cured/Discharged/Migrated'].max().sort_values(ascending=False).head()
plt.bar(top5.index, top5.values, label="Recovered cases")

plt.title("COVID Recovered Cases")
plt.ylabel("Number of Cases")
plt.xlabel("State")
plt.legend(title="Case Type")
plt.show()
```

```python
top5 = df.groupby('Name of State / UT')['Cured/Discharged/Migrated'].max().sort_values(ascending=False).head()

plt.pie(
    top5.values,       # numeric values
    labels=top5.index, # state names
    autopct="%1.1f%%",
    startangle=90
)
plt.title("Cured COVID-19 Cases")
plt.show()
```

```python
top5 = df.groupby('Name of State / UT')['Death'].max().sort_values(ascending=False).head()

plt.pie(
    top5.values,       # numeric values
    labels=top5.index, # state names
    autopct="%1.1f%%",
    startangle=90
)
plt.title("Death COVID-19 Cases")
plt.show()
```

```python
plt.plot(df["Date"],df["Cured/Discharged/Migrated"],marker = "*",label="Recovered")
plt.plot(df["Date"],df["Death"],marker="o",label="Death")
plt.title("Day wise cases")
plt.xlabel("Total Confirmed cases")
plt.ylabel("")
plt.legend()
```
